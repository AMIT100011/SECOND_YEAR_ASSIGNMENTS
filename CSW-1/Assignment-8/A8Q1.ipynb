```python
import numpy as np
v1 = np.random.randint(1,50,10)
v2 = np.random.randint(1,50,10)
mx1 = np.max(v1)
mn1 = np.min(v1)
mx2 = np.max(v2)
mn2 = np.min(v2)
print("Array:",v1)
print("Array:",v2)
print("Max1:",mx1)
print("Min1:",mn1)
print("Max2:",mx2)
print("Min2:",mn2)
ele1 = ((v1<mx1) & (v1>mn1))
ele2 = ((v2<mx2) & (v2>mn2))

print("Elements greater than max and less than min:",v1[ele1])
print("Elements greater than max and less than min:",v2[ele2])
```
