```python
import numpy as np

v1 = np.random.randint(1, 10, (1, 3))
v2 = np.random.randint(1, 10, (1, 3))

len_v1 = np.linalg.norm(v1)
len_v2 = np.linalg.norm(v2)

v1Norm = v1 / len_v1
v2Norm = v2 / len_v2

dot = np.dot(v1Norm, v2Norm.T)
angle_rad = np.arccos(dot)
angle_deg = np.degrees(angle_rad)

print("Length v1:", len_v1)
print("Length v2:", len_v2)
print("Angle (rad):", angle_rad[0][0])
print("Angle (deg):", angle_deg[0][0])
```
