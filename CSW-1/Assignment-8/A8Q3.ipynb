```python
import numpy as np
import time
import random

start1 = time.time()
np.random.random((1000, 1000))
end1 = time.time()

py_time = end1 - start1

start = time.time()
arr = [[random.random() for _ in range(1000)] for _ in range(1000)]
end = time.time()
arr_time = end - start

print(f"Numpy time: {py_time}")
print(f"array time: {arr_time}")
print("Numpy" if py_time < arr_time else "array")

```
