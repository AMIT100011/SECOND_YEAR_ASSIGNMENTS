```python
import numpy as np

temp = np.array([28, 32, 35, 37, 30, 29, 34, 36, 31, 33, 38, 27, 40, 34])
print("temperature above 35 :", temp[temp>=35])
print("Count of temperature above 35 :", np.sum(temp>=35))
temp[temp<30] = 30
print("Modified temperature array :", temp)
comfortable = (temp>=30) & (temp<=34)
print("Comfortable temperature range (30-34) :", temp[comfortable])
temp[temp>=35] += 2
print("Temperature array after increasing values >=35 by 2 :", temp)
```
