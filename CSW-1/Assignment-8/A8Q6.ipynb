```python
import numpy as np
def generate():
  return np.random.randint(1,1000,999)
def calculate(arr):
  m = np.mean(arr)
  s = np.std(arr)
  return {
     "1_std": int(np.sum((arr >= m-s) & (arr <= m+s))),
     "2_std": np.sum((arr >= m-2*s) & (arr <= m+2*s)),
     "3_std": np.sum((arr >= m-3*s) & (arr <= m+3*s))
  }
data = generate()
result = calculate(data)
print(result)
```
