```python
import numpy as np

r = int(input("Enter rows: "))
c = int(input("Enter columns: "))

trials = np.zeros(r, dtype=int)

# Hidden matrix
hidden = np.random.randint(1, 20, size=(r, c))

print("\nHidden matrix (for testing):")
print(hidden)

for i in range(r):
    print(f"\nGuessing for Row {i + 1}")
    while True:
        trials[i] += 1
        guess = input(
            f"Enter your {c} guesses separated by space: "
        ).split()

        # convert input to integer numpy array
        guess = np.array(guess, dtype=int)

        if np.array_equal(guess, hidden[i]):
            print("‚úÖ Correct!")
            break
        else:
            print("‚ùå Incorrect, try again.")

print("\nüéØ Number of trials for each row:")
for i in range(r):
    print(f"Row {i + 1}: {trials[i]} trials")
```
