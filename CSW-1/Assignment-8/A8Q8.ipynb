```python
import numpy as np

mat = np.array([np.random.randint(i+1, i+9, 5) for i in range(4)])

row_mean = np.mean(mat, axis=1)
col_mean = np.mean(mat, axis=0)

row_std = np.std(mat, axis=1)
col_std = np.std(mat, axis=0)

result = {
    "mean": "row-wise" if np.mean(row_mean) > np.mean(col_mean) else "column-wise",
    "std": "row-wise" if np.mean(row_std) > np.mean(col_std) else "column-wise"
}

print(result)
```
