```python
import numpy as np

arr = np.random.rand(4,4,4)
norm_axis0 = arr / np.linalg.norm(arr, axis=0, keepdims=True)
norm_axis1 = arr / np.linalg.norm(arr, axis=1, keepdims=True)
norm_axis2 = arr / np.linalg.norm(arr, axis=2, keepdims=True)
print("Normalized along axis 0:\n", norm_axis0)
print("Normalized along axis 1:\n", norm_axis1)
print("Normalized along axis 2:\n", norm_axis2)
```
