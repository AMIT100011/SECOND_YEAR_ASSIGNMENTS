```python
import numpy as np


arr = np.linspace(0, np.pi, 10)

sin_values = np.sin(arr)
cos_values = np.cos(arr)
sqrt_values = np.sqrt(arr)

print(f"{'Value':>8} {'Sin':>10} {'Cos':>10} {'Sqrt':>10}")
print("-" * 44)

for i in range(len(arr)):
    print(f"{arr[i]:8.4f} {sin_values[i]:10.4f} {cos_values[i]:10.4f} {sqrt_values[i]:10.4f}")
```
