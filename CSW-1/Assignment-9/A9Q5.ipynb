```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-20, 25, 100)
y1 = x**3 - 15*x**2 + 25
y2 = 2*x**2 - 10*x + 5
plt.plot(x, y1, label="y1", linestyle='-')
plt.plot(x, y2, label="y2", linestyle='--')
plt.title("Polynomial Graph")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
