```python
import matplotlib.pyplot as plt

# Create figure
plt.figure(figsize=(6, 6))

# Draw X and Y axes
plt.axhline(0, color='black', linewidth=1.5)
plt.axvline(0, color='black', linewidth=1.5)

# -------- LEFT RED SHAPE --------
# Upper triangle
x1 = [-1.2, -0.8, -1.0, -1.2]
y1 = [1.2, 1.2, 0, 1.2]

# Lower triangle
x2 = [-1.2, -0.8, -1.0, -1.2]
y2 = [-1.2, -1.2, 0, -1.2]

plt.plot(x1, y1, color='red', linewidth=2)
plt.plot(x2, y2, color='red', linewidth=2)

# -------- RIGHT GREEN SHAPE --------
# Upper triangle
x3 = [0.8, 1.2, 1.0, 0.8]
y3 = [1.2, 1.2, 0, 1.2]

# Lower triangle
x4 = [0.8, 1.2, 1.0, 0.8]
y4 = [-1.2, -1.2, 0, -1.2]

plt.plot(x3, y3, color='lime', linewidth=2)
plt.plot(x4, y4, color='lime', linewidth=2)

# Formatting
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.gca().set_aspect('equal', adjustable='box')
plt.axis('off')

# Show plot
plt.show()
```
