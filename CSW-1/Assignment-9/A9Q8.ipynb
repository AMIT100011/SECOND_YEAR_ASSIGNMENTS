```python
import numpy as np
import matplotlib.pyplot as plt
v1 = np.array([3, 4])
v2 = np.array([4, 3])
v1 = v1 / np.linalg.norm(v1)
v2 = v2 / np.linalg.norm(v2)
plt.quiver(0, 0, v1[0], v1[1], color='g', scale=1, label="v1")
plt.quiver(0, 0, v2[0], v2[1], color='r', scale=1, label="v2")
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.gca().set_aspect('equal')

plt.legend()
plt.show()
```
